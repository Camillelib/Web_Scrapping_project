## Scrap for all pages

```python
import time

for i in range(n+1):
    url=f'https://www.carrefour.fr/promotions?noRedirect=0&page={i}'
    html = r.get(url).content
    soup = BeautifulSoup(html)
    product_name=[i.text.replace('\n','') for i in soup.select('ul h2')]
    product_price_total = soup.find_all('div',{'class':'ds-body-text ds-body-text--size-s ds-body-text--color-standard-3 product-card-per-unit-label'});
    product_price_total=[i.text.replace('\n','') for i in product_price_total]
    product_quantity = soup.find_all('div',{'class':'ds-body-text ds-body-text--size-s ds-body-text--color-standard-3 product-card-packaging'});
    product_quantity=[i.text.replace('\n','') for i in product_quantity]
    product_price=[i.text.replace('\n','') for i in soup.select('div.product-card-price')]
    promotion_where=[i.text.replace('\n','') for i in soup.select('div.ds-product-card--vertical-highlightlabel')]
    promotion_type=[i.text.replace('\n','') for i in soup.select('div.promotion-description__labels')]
    dct={'Product name':product_name, 'Unit price':product_price, 'Total price':product_price_total, 'Quantity': product_quantity, 'Promotion Type':promotion_type, 'Promotion comes from':promotion_where }
    df=pd.DataFrame(dct)
    time.sleep(2)
    print(i)
```
